## Letra D)

A folha de pagamento total desta empresa, sendo que cada funcionário terá:

*   Desconto: 2% do Salário Bruto, se o funcionário tiver plano de saúde.

*   Acréscimo: Vale Alimentação de 1% por dependente.

```
# LIQUIDO = BRUTO - DESCONTOS + BENEFICIOS
bruto = sum(dataset1[:,1])
round(bruto, 2)
```

```
descontos = 0
for indice, funcionario in enumerate(dataset1[:]):
  if funcionario[2] == 1:
    descontos += funcionario[1]
round(descontos, 2)
```

```
beneficios = 0
for funcionario in dataset1[:]:
    beneficios += funcionario[1] * 0.01 * funcionario[3]
round(beneficios, 2)
```

```
liquido = bruto - descontos + beneficios
print(f'A folha de pagamento foi de: {liquido: .2f}')
```

## Letra A)

O Gerente de CRM (Customer Relationship Management - Gerente de Relacionamento com o Cliente) da Empresa quer fazer uma Distribuição de Cartão de Fidelidade, de acordo com a tabela de compras.

Qual a quantidade: Absoluta e relativa (%) de clientes por cartão.

<center>

image.png

```
clientesBronze = 0
clientesPrata = 0
clientesOuro = 0
clientesDiamante = 0 

for cliente in dataset3[:]:
  if cliente[1] <= 100.00:
    clientesBronze += 1
  elif cliente[1] <= 500.00:
    clientesPrata += 1
  elif cliente[1]  <= 1000.00:
    clientesOuro += 1
  elif cliente[1] > 1000.00:
    clientesDiamante += 1
totalClientes = len(dataset3[:])
```

```
print('A frequência absoluta e relativa são respectivamente: ')
print(f'Clientes bronze:   {clientesBronze}   e {clientesBronze / totalClientes}')
print(f'Clientes prata:    {clientesPrata}   e {clientesPrata / totalClientes}')
print(f'Clientes ouro:     {clientesOuro}  e {clientesOuro / totalClientes}')
print(f'Clientes diamante: {clientesDiamante} e {clientesDiamante / totalClientes}')
```

## Letra B)

Custo da campanha CRM desta empresa por grupo e total (Custo de confecção unitário do cartão: R$ 3.87).

```
custoBronze = clientesBronze * 3.87
custoPrata = clientesPrata * 3.87
custoOuro = clientesOuro * 3.87
custoDiamante = clientesDiamante * 3.87
```

```
print(f'Os custos serão de:')
print(f'BRONZE:   R${custoBronze}')
print(f'PRATA:    R${custoPrata}')
print(f'OURO:     R${custoOuro}')
print(f'DIAMANTE: R${custoDiamante}')
```

## Letra C) 

O perfil dos clientes por sexo, com os seguintes dados;

*   Mínimo gasto

*   Máximo gasto

*   Média de gastos

```
gastos = dataset3[:, 1] 
sexo = dataset3[:, 2]
```

```
masculino = (sexo == 0)
feminino = (sexo == 1)
```

```
minGastoMasc = np.argmin(gastos[masculino]) 
minGastoFem = np.argmin(gastos[feminino]) 

menorValorM = gastos[masculino][minGastoMasc]
menorValorF = gastos[feminino][minGastoFem]
```

```
maxGastoMasc = np.argmax(gastos[masculino]) 
maxGastoFem = np.argmax(gastos[feminino]) 

maiorValorM = gastos[masculino][maxGastoMasc]
maiorValorF = gastos[feminino][maxGastoFem]
```

```
avgGastoMasc = sum(gastos[masculino][:]) / len(gastos[masculino])
avgGastoFem = sum(gastos[feminino][:]) / len(gastos[feminino])
```

```
print(f'Mínimo gasto F: {menorValorF: .2f}')
print(f'Mínimo gasto M: {menorValorM : .2f}')
print('')
print(f'Máximo gasto F: {maiorValorF: .2f}')
print(f'Máximo gasto M: {maiorValorM: .2f}')
print('')
print(f'Média gasto F:  {avgGastoFem: .2f}')
print(f'Média gasto M:  {avgGastoMasc: .2f}')
```

## Letra E) 

O perfil dos clientes por tipo de cartão, com os seguintes dados;

*   Mínimo gasto

*   Máximo gasto

*   Média de gastos

```
gastos = dataset3[:,1]
```

```
indexBronze = dataset3[:,1] <= 100
indexPrata = (dataset3[:,1] > 100) & (dataset3[:,1] <= 500)
indexOuro = (dataset3[:,1] > 500) & (dataset3[:,1] <= 1000)
indexDiamante = dataset3[:,1] > 1000
```

```
minGastoB = np.argmin(gastos[indexBronze]) 
minGastoP = np.argmin(gastos[indexPrata]) 
minGastoO = np.argmin(gastos[indexOuro]) 
minGastoD = np.argmin(gastos[indexDiamante]) 

menorValorB = gastos[indexBronze][minGastoB]
menorValorP = gastos[indexPrata][minGastoP]
menorValorO = gastos[indexOuro][minGastoO]
menorValorD = gastos[indexDiamante][minGastoD]
```

```
maxGastoB = np.argmax(gastos[indexBronze]) 
maxGastoP = np.argmax(gastos[indexPrata]) 
maxGastoO = np.argmax(gastos[indexOuro]) 
maxGastoD = np.argmax(gastos[indexDiamante]) 

maiorValorB = gastos[indexBronze][maxGastoB]
maiorValorP = gastos[indexPrata][maxGastoP]
maiorValorO = gastos[indexOuro][maxGastoO]
maiorValorD = gastos[indexDiamante][maxGastoD]
```

```
avgGastoB = sum(gastos[indexBronze][:]) / len(gastos[indexBronze])
avgGastoP = sum(gastos[indexPrata][:]) / len(gastos[indexPrata])
avgGastoO = sum(gastos[indexOuro][:]) / len(gastos[indexOuro])
avgGastoD = sum(gastos[indexDiamante][:]) / len(gastos[indexDiamante])
```

```
print(f'Mínimo gasto B: {menorValorB: .2f}')
print(f'Mínimo gasto P: {menorValorP : .2f}')
print(f'Mínimo gasto O: {menorValorO: .2f}')
print(f'Mínimo gasto D: {menorValorD : .2f}')
print('')
print(f'Máximo gasto B: {maiorValorB: .2f}')
print(f'Máximo gasto P: {maiorValorP: .2f}')
print(f'Máximo gasto O: {maiorValorO: .2f}')
print(f'Máximo gasto D: {maiorValorD: .2f}')
print('')
print(f'Média gasto B:  {avgGastoB: .2f}')
print(f'Média gasto P:  {avgGastoP: .2f}')
print(f'Média gasto O:  {avgGastoO: .2f}')
print(f'Média gasto D:  {avgGastoD: .2f}')
```
